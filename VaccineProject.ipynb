```python
#Creates lists with all of the outbreak locations and tweet locations

import csv

with open('outbreakLocationData.csv', 'r') as f:
    reader = csv.reader(f)
    obLocs0 = list(reader)
    
with open('twitterLocationData.csv', 'r') as f:
    reader = csv.reader(f)
    twtLocs0 = list(reader)
    
#fixes lists so they are 1D as opposed to 2D (like the parsing module makes them)

obLocs = []
twtLocs = []

for ob in obLocs0:
    obLocs.append(ob[0])
    
for twt in twtLocs0:
    twtLocs.append(twt[0])
    
```

```python
#makes dictionaries counting the number of occrances of each location for outbreaks and tweets

from collections import defaultdict
    
twtDict = defaultdict(lambda: 0)
obDict = defaultdict(lambda: 0)
    
for twt in twtLocs:
    twtDict[twt] += 1
    
for ob in obLocs:
    obDict[ob] += 1

#turns dictionaries into a list of tuples

twtList = []
twtName = []
twtCount = []

obList = []
obName = []
obCount = []

for val in twtDict:
    twtList.append((twtDict[val], val))
    
for val in obDict:
    obList.append((obDict[val], val))
    
twtList.sort(reverse = True)
obList.sort(reverse = True)
    
for i in twtList:
    twtName.append(i[1])
    twtCount.append(i[0])
    
for i in obList:
    obName.append(i[1])
    obCount.append(i[0])
```
